```python
from google.colab import files
import pandas as pd
import matplotlib as plt
import numpy as np

uploaded = files.upload()
file_name = next(iter(uploaded))
```

```python
data = pd.read_csv(file_name,sep=',')
data
```

```python
from datetime import datetime,timedelta
from dateutil import parser

def transform_date(date):
    try:
        fecha_objeto = parser.parse(date)
        nueva_cadena_fecha = fecha_objeto.strftime('%Y-%m-%d')
    except ValueError:
        fecha_objeto = datetime.strptime(date, '%d-%m-%y')
        nueva_cadena_fecha = fecha_objeto.strftime('%Y-%m-%d')
    return nueva_cadena_fecha
```

```python
data['birthdate']=data['birthdate'].apply(transform_date)
```

```python
data['birthdate'] = pd.to_datetime(data['birthdate'])
```

```python
indices_mayores = data.index[data['birthdate'] > datetime.now()]
data.loc[indices_mayores, 'birthdate'] -= pd.DateOffset(years=100)
```

```python
data['Age']= (datetime.now() - data['birthdate']).astype('<m8[Y]')
```
